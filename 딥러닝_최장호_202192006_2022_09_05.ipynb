```python
def sigmoid(x):
    return 1/(1+np.exp(-x))

xx = np.linspace(-10, 10, 100)
plt.plot(xx, sigmoid(xx))

plt.xlabel("$x$")
plt.ylabel("$\sigma(x)$")
plt.show()
```

```python
def D_sigmoid(x):
    return sigmoid(x) * (1 - sigmoid(x))

xx = np.linspace(-10, 10, 100)
plt.plot(xx, D_sigmoid(xx))

plt.show()
```

## tanh

```python
def tanh(x):
    return(np.exp(x) - np.exp(-x)) / (np.exp(x)+np.exp(-x))

xx = np.linspace(-10, 10, 100)
plt.plot(xx, tanh(xx))

plt.xlabel("$x$")
plt.ylabel("$\sigma(x)$")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt


def tanh(x):
    return(np.exp(x) - np.exp(-x)) / (np.exp(x)+np.exp(-x))

x = np.linspace(-10, 10, 100)
plt.plot(x, (1+tanh(x))*(1-tanh(x)))


plt.show()
```
